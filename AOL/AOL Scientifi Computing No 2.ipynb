```python
import sympy as sp
import math
```

```python
# Define the variable x
x = sp.Symbol('x')

# Compute the fourth-order Taylor expansion for sin(x)
sin_taylor = sp.series(sp.sin(x), x, 0, 5).removeO()

# Compute the fourth-order Taylor expansion for cos(x)
cos_taylor = sp.series(sp.cos(x), x, 0, 5).removeO()

# Compute the fourth-order Taylor expansion for sin(x)cos(x)
sin_cos_taylor = sp.series(sp.sin(x)*sp.cos(x), x, 0, 5).removeO()

# Print the Taylor expansions
print(f"sin(x) Taylor expansion: {sin_taylor}")
print(f"cos(x) Taylor expansion: {cos_taylor}")
print(f"sin(x)*cos(x) Taylor expansion: {sin_cos_taylor}")
```

```python
# Define the variable x and the value for x = π/2
x = sp.Symbol('x')
x_val = math.pi / 2

# Compute the fourth-order Taylor expansion for sin(x) and cos(x) separately
sin_taylor = sp.series(sp.sin(x), x, 0, 5).removeO()
cos_taylor = sp.series(sp.cos(x), x, 0, 5).removeO()

# Compute the fourth-order Taylor expansion for sin(x)cos(x) by multiplying the Taylor expansions
sin_cos_taylor_mult = sin_taylor * cos_taylor

# Compute the fourth-order Taylor expansion for sin(x)cos(x) by plugging in x
sin_cos_taylor_plug = sp.series(sp.sin(x)*sp.cos(x), x, 0, 5).removeO().subs(x, x_val)

# Calculate the error for each approach
error_mult = abs(sin_cos_taylor_mult.subs(x, x_val) - math.sin(x_val)*math.cos(x_val))
error_plug = abs(sin_cos_taylor_plug - math.sin(x_val)*math.cos(x_val))

# Print the errors
print(f"Error using multiplication approach: {error_mult}")
print(f"Error using plugging in x approach: {error_plug}")
```

```python
# Define the variable x and the value for x = π/4
x = sp.Symbol('x')
x_val = math.pi / 4

# Define the function cos(x)
cos_func = sp.cos(x)

# Compute the Taylor series expansion for cos(x) around x = 0
taylor_expansion = sp.series(cos_func, x, 0, 5).removeO()

# Evaluate the Taylor series at x = π/4
taylor_approximation = taylor_expansion.subs(x, x_val)

# Calculate the truncation error bound for the Taylor series
remainder_term = sp.series(cos_func, x, 0, 6).removeO()
error_bound = abs(remainder_term.subs(x, x_val))

# Print the Taylor approximation and truncation error bound
print(f"Taylor Approximation: {taylor_approximation}")
print(f"Truncation Error Bound: {error_bound}")
```
